```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generate line points
def line_gen(A,B):
    len =10
    dim = A.shape[0]
    x_AB = np.zeros((dim,len))
    lam_1 = np.linspace(0,1,len)
    for i in range(len):
        temp1 = A + lam_1[i]*(B-A)
        x_AB[:,i]= temp1.T
    return x_AB
```

```python
def area_parallelogram(a=1):

  # defining the equation matrices
  A = np.array([[12, -5], [5, -12]])
  B = np.array([[5, -12], [12, -5]])
  C = np.array([7*a, -7*a])
  D = np.array([7*a, -126*a])
  E = np.array([-7*a, 126*a])
  F = np.array([-126*a, 126*a])

  # finding the vertices of ||gm
  M = np.linalg.inv(A) @ C
  N = np.linalg.inv(A) @ D
  P = np.linalg.inv(B) @ E
  O = np.linalg.inv(B) @ F
  # print(M, N, O)

  # area equals the L2 norm of the cross product
  area = np.linalg.norm(np.cross(N-M, P-M))


  # plotting the parallelogram

  # line generation
  x_MN = line_gen(M, N)
  x_NO = line_gen(N, O)
  x_OP = line_gen(O, P)
  x_PM = line_gen(P, M)


  plt.plot(x_MN[0,:],x_MN[1,:],label='$12x-5y-7=0$')
  plt.plot(x_NO[0,:],x_NO[1,:],label='$5x-12y+126=0$')
  plt.plot(x_OP[0,:],x_OP[1,:],label='$12x-5y-126=0$')
  plt.plot(x_PM[0,:],x_PM[1,:],label='$5x-12y+7=0$')

  plt.legend(loc='upper left')
  plt.annotate("M", M)
  plt.annotate("N", N)
  plt.annotate("O", O)
  plt.annotate("P", P)

  plt.ylabel('$y-axis$')
  plt.xlabel('$x-axis$')
  plt.grid() 
  name = 'parallelogram_'+str(a)+'.jpg'
  plt.savefig(name, dpi=200)
  plt.figure(figsize=(16, 18), dpi=200)
  plt.show()


  return area
```

```python
print(area_parallelogram(1), area_parallelogram(2))
```
